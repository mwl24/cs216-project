```python
# import libraries

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
players = pd.read_csv("data/players.csv")
print(players.shape)
players.rename(columns={'_id':'player_id'}, inplace=True)
players.head()
```

```python
salaries = pd.read_csv("data/salaries_1985to2018.csv")
print(salaries.shape)
salaries.head()
```

```python
agg_salaries = salaries.groupby('player_id').sum()
agg_salaries.head()
```

```python
df = players.merge(agg_salaries, on = 'player_id')
df
```
